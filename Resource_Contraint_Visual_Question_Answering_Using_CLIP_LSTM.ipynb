```python
@lru_cache()
def default_bpe():
    return "experiment/model/bpe_simple_vocab_16e6.txt.gz"


@lru_cache()
def bytes_to_unicode():
    """
    Returns list of utf-8 byte and a corresponding list of unicode strings.
    The reversible bpe codes work on unicode strings.
    This means you need a large # of unicode characters in your vocab if you
    want to avoid UNKs. When you're at something like a 10B token dataset you
    end up needing around 5K for decent coverage. This is a signficant
    percentage of your normal, say, 32K bpe vocab. To avoid that, we want lookup
    tables between utf-8 bytes and unicode strings. And avoids mapping to
    whitespace/control characters the bpe code barfs on.
    """
    bs = (
        list(range(ord("!"), ord("~") + 1))
        + list(range(ord("¡"), ord("¬") + 1))
        + list(range(ord("®"), ord("ÿ") + 1))
    )
    cs = bs[:]
    n = 0
    for b in range(2**8):
        if b not in bs:
            bs.append(b)
            cs.append(2**8 + n)
            n += 1
    cs = [chr(n) for n in cs]
    return dict(zip(bs, cs))


def get_pairs(word):
    """Return set of symbol pairs in a word.
    Word is represented as tuple of symbols (symbols being variable-length
    strings).
    """
    pairs = set()
    prev_char = word[0]
    for char in word[1:]:
        pairs.add((prev_char, char))
        prev_char = char
    return pairs


def basic_clean(text):
    text = ftfy.fix_text(text)
    text = html.unescape(html.unescape(text))
    return text.strip()


def whitespace_clean(text):
    text = re.sub(r"\s+", " ", text)
    text = text.strip()
    return text
```

```python
class SimpleTokenizer(object):
    def __init__(self, bpe_path: str = default_bpe()):
        self.byte_encoder = bytes_to_unicode()
        self.byte_decoder = {v: k for k, v in self.byte_encoder.items()}
        merges = gzip.open(bpe_path).read().decode("utf-8").split("\n")
        merges = merges[1 : 49152 - 256 - 2 + 1]
        merges = [tuple(merge.split()) for merge in merges]
        vocab = list(bytes_to_unicode().values())
        vocab = vocab + [v + "</w>" for v in vocab]
        for merge in merges:
            vocab.append("".join(merge))
        vocab.extend(["<|startoftext|>", "<|endoftext|>"])
        self.encoder = dict(zip(vocab, range(len(vocab))))
        self.decoder = {v: k for k, v in self.encoder.items()}
        self.bpe_ranks = dict(zip(merges, range(len(merges))))
        self.cache = {
            "<|startoftext|>": "<|startoftext|>",
            "<|endoftext|>": "<|endoftext|>",
        }
        self.pat = re.compile(
            r"""<\|startoftext\|>|<\|endoftext\|>|'s|'t|'re|'ve|'m|'ll|'d|[\p{L}]+|[\p{N}]|[^\s\p{L}\p{N}]+""",
            re.IGNORECASE,
        )

    def bpe(self, token):
        if token in self.cache:
            return self.cache[token]
        word = tuple(token[:-1]) + (token[-1] + "</w>",)
        pairs = get_pairs(word)

        if not pairs:
            return token + "</w>"

        while True:
            bigram = min(pairs, key=lambda pair: self.bpe_ranks.get(pair, float("inf")))
            if bigram not in self.bpe_ranks:
                break
            first, second = bigram
            new_word = []
            i = 0
            while i < len(word):
                try:
                    j = word.index(first, i)
                    new_word.extend(word[i:j])
                    i = j
                except:
                    new_word.extend(word[i:])
                    break

                if word[i] == first and i < len(word) - 1 and word[i + 1] == second:
                    new_word.append(first + second)
                    i += 2
                else:
                    new_word.append(word[i])
                    i += 1
            new_word = tuple(new_word)
            word = new_word
            if len(word) == 1:
                break
            else:
                pairs = get_pairs(word)
        word = " ".join(word)
        self.cache[token] = word
        return word

    def encode(self, text):
        bpe_tokens = []
        text = whitespace_clean(basic_clean(text)).lower()
        for token in re.findall(self.pat, text):
            token = "".join(self.byte_encoder[b] for b in token.encode("utf-8"))
            bpe_tokens.extend(
                self.encoder[bpe_token] for bpe_token in self.bpe(token).split(" ")
            )
        return bpe_tokens

    def decode(self, tokens):
        text = "".join([self.decoder[token] for token in tokens])
        text = (
            bytearray([self.byte_decoder[c] for c in text])
            .decode("utf-8", errors="replace")
            .replace("</w>", " ")
        )
        return text
```

```python
SimpleTokenizer = SimpleTokenizer()
```

```python
def tokenize(texts, context_length = MAX_SEQUENCE, truncate = True):
    if isinstance(texts, str):
        texts = [texts]

    sot_token = SimpleTokenizer.encoder[START_TOKEN]
    eot_token = SimpleTokenizer.encoder[END_TOKEN]
    all_tokens = [
        [sot_token] + SimpleTokenizer.encode(text) + [eot_token]
        for text in texts
    ]

    result = torch.zeros(len(all_tokens), context_length, dtype=torch.long)

    for i, tokens in enumerate(all_tokens):
        if len(tokens) > context_length:
            if truncate:
                tokens = tokens[:context_length]
                tokens[-1] = eot_token
            else:
                raise RuntimeError(
                    f"Input {texts[i]} is too long for context length {context_length}"
                )
        result[i, : len(tokens)] = torch.tensor(tokens)

    return result
```

#### **VQA Dataset**

```python
class VizWizDataset(data.Dataset):
    def __init__(self, split, transform=TRANSFORM):
        assert split in ["train", "val", "test", "inference"]
        if split in ["train", "val"]:
            path_to_annotation = "experiment/dataset/train.json"
            self.root_dir = "experiment/dataset/train"
        elif split == "test":
            path_to_annotation = "experiment/dataset/val.json"
            self.root_dir = "experiment/dataset/val"
        elif split == "inference":
            path_to_annotation = "experiment/dataset/test.json"
            self.root_dir = "experiment/dataset/test"
        else:
            raise TypeError(f"{split}")

        with open(path_to_annotation) as f:
            json_data = json.load(f)

        self.dataset = list()

        for jdat in json_data:
            try:
                if jdat['answer_type'] == "unanswerable":
                    self.dataset.append(
                        (
                            jdat["image"],
                            jdat["question"].lower(),
                            'unanswerable'
                        )
                    )
                    continue

                if jdat['answer_type'] == "yes/no":
                    sentiment = None
                    if 'yes' in jdat["answers"][0]['answer'].lower():
                        sentiment = "yes"
                    else:
                        sentiment = "no"
                    self.dataset.append(
                        (
                            jdat["image"],
                            jdat["question"].lower(),
                            sentiment
                        )
                    )
                    continue

                answer_candidate = dict()
                answers = list(jdat["answers"])
                random.shuffle(answers)
                for ans in answers:
                    if ans['answer'].lower() in answer_candidate.keys():
                        answer_candidate[ans["answer"].lower()] += 1
                    else:
                        answer_candidate[ans["answer"].lower()] = 1

                best_answer_count = 0
                best_answer_candidate = None
                for k, v in answer_candidate.items():
                    if v > best_answer_count:
                        best_answer_candidate = k
                        best_answer_count = v
                self.dataset.append(
                    (
                        jdat["image"],
                        jdat["question"].lower(),
                        best_answer_candidate
                    )
                )
            except:
                self.dataset.append((jdat["image"], jdat["question"].lower()))

        if split == "train":
            random.seed(SEED)
            random.shuffle(self.dataset)
            self.dataset = self.dataset[: int(0.99 * len(self.dataset))]
        elif split == "val":
            random.seed(SEED)
            random.shuffle(self.dataset)
            self.dataset = self.dataset[int(0.99 * len(self.dataset)) :]

        self.split = split
        self.transform = transform

    def __len__(self):
        return len(self.dataset)

    def get_raw_question(self, index):
        assert self.split == "inference"
        return self.dataset[index][1]

    def get_raw_image(self, index):
        assert self.split == "inference"
        return Image.open(
            os.path.join(
                self.root_dir,
                self.dataset[index][0],
            )
        )

    def get_inference_data(self, index):
        assert self.split == "inference"
        image = Image.open(
            os.path.join(
                self.root_dir,
                self.dataset[index][0],
            )
        )
        question = self.dataset[index][1]

        if self.transform is not None:
            image = self.transform(image)

        question = tokenize(question)[0]

        return image, question.long()

    def __getitem__(self, index):
        assert self.split != "inference"
        image = Image.open(
            os.path.join(
                self.root_dir,
                self.dataset[index][0],
            )
        )
        question = self.dataset[index][1]
        answer = self.dataset[index][2]

        if self.transform is not None:
            image = self.transform(image)

        question = tokenize(question)[0]
        answer = tokenize(answer)[0]

        return image, question.long(), answer.long()
```

#### **Collate Function**

```python
class CollateFunction(object):
    def __init__(self, pad_idx=0):
        self.pad_idx = pad_idx

    def __call__(self, batch):
        images = [item[0].unsqueeze(0) for item in batch]
        images = torch.cat(images, dim=0)

        questions = [item[1].unsqueeze(0) for item in batch]
        questions = torch.cat(questions, dim=0)

        answers = [
            item[2] for item in batch
        ] + [torch.randn(MAX_SEQUENCE + 1)]
        answers = pad_sequence(
            answers,
            batch_first=True,
            padding_value=self.pad_idx,
        )[:-1, ...]
        return images, questions.long(), answers.long()
```

### **Module**

```python
def build_attention_mask():
    # lazily create causal attention mask, with full attention between the
    # vision tokens pytorch uses additive attention mask; fill with -inf
    mask = torch.empty(MAX_SEQUENCE, MAX_SEQUENCE)
    mask.fill_(float("-inf"))
    mask.triu_(1)  # zero out the lower diagonal
    return mask
```

```python
class QuickGELU(nn.Module):
    def forward(self, x: torch.Tensor):
        return x * torch.sigmoid(1.702 * x)
```

```python
class ResidualAttentionBlock(nn.Module):
    def __init__(self, d_model: int, n_head: int, attn_mask: torch.Tensor = None):
        super().__init__()

        self.attn = nn.MultiheadAttention(d_model, n_head)
        self.ln_1 = nn.LayerNorm(d_model)
        self.mlp = nn.Sequential(
            collections.OrderedDict(
                [
                    ("c_fc", nn.Linear(d_model, d_model * 4)),
                    ("gelu", QuickGELU()),
                    ("c_proj", nn.Linear(d_model * 4, d_model)),
                ]
            )
        )
        self.ln_2 = nn.LayerNorm(d_model)
        self.attn_mask = attn_mask

    def attention(self, x: torch.Tensor):
        self.attn_mask = (
            self.attn_mask.to(dtype=x.dtype, device=x.device)
            if self.attn_mask is not None
            else None
        )
        return self.attn(x, x, x, need_weights=False, attn_mask=self.attn_mask)[0]

    def forward(self, x: torch.Tensor):
        x = x + self.attention(self.ln_1(x))
        x = x + self.mlp(self.ln_2(x))
        return x
```

```python
class Transformer(nn.Module):
    def __init__(
        self, width: int, layers: int, heads: int, attn_mask: torch.Tensor = None
    ):
        super().__init__()
        self.width = width
        self.layers = layers
        self.resblocks = nn.Sequential(
            *[ResidualAttentionBlock(width, heads, attn_mask) for _ in range(layers)]
        )

    def forward(self, x: torch.Tensor):
        return self.resblocks(x)
```

```python
class VisionTransformer(nn.Module):
    def __init__(
        self,
        input_resolution: int = IMAGE_SIZE,
        patch_size: int = 32,
        width: int = 768,
        layers: int = 12,
        heads: int = 12,
        output_dim: int = HIDDEN_DIM,
    ):
        super().__init__()
        self.input_resolution = input_resolution
        self.output_dim = output_dim
        self.conv1 = nn.Conv2d(
            in_channels=3,
            out_channels=width,
            kernel_size=patch_size,
            stride=patch_size,
            bias=False,
        )

        scale = width**-0.5
        self.class_embedding = nn.Parameter(scale * torch.randn(width))
        self.positional_embedding = nn.Parameter(
            scale * torch.randn((input_resolution // patch_size) ** 2 + 1, width)
        )
        self.ln_pre = nn.LayerNorm(width)

        self.transformer = Transformer(width, layers, heads)

        self.ln_post = nn.LayerNorm(width)
        self.proj = nn.Parameter(scale * torch.randn(width, output_dim))

    def forward(self, x: torch.Tensor):
        x = self.conv1(x)  # shape = [*, width, grid, grid]
        x = x.reshape(x.shape[0], x.shape[1], -1)  # shape = [*, width, grid ** 2]
        x = x.permute(0, 2, 1)  # shape = [*, grid ** 2, width]
        x = torch.cat(
            [
                self.class_embedding.to(x.dtype)
                + torch.zeros(
                    x.shape[0], 1, x.shape[-1], dtype=x.dtype, device=x.device
                ),
                x,
            ],
            dim=1,
        )  # shape = [*, grid ** 2 + 1, width]
        x = x + self.positional_embedding.to(x.dtype)
        x = self.ln_pre(x)

        x = x.permute(1, 0, 2)  # NLD -> LND
        x = self.transformer(x)
        x = x.permute(1, 0, 2)  # LND -> NLD

        x = self.ln_post(x[:, 0, :])

        if self.proj is not None:
            x = x @ self.proj

        return x
```

```python
class Decoder(nn.Module):
    def __init__(self):
        super().__init__()

        self.token_embedding = nn.Embedding(VOCAB_SIZE, HIDDEN_DIM)
        self.token_embedding.weight.data = clip.token_embedding.weight.data.to(
            torch.float32
        ).clone()
        self.token_embedding.weight.requires_grad = UNFREEZE_PARAM

        self.lstm = nn.LSTM(HIDDEN_DIM, HIDDEN_DIM, NUM_LAYER, batch_first=True)

        self.in_projection = nn.Linear(2 * HIDDEN_DIM, HIDDEN_DIM)
        self.out_projection = nn.Linear(HIDDEN_DIM, VOCAB_SIZE)

    def forward(self, x, y, z):
        x = self.token_embedding(x)
        x = x * self.in_projection(torch.cat([y, z], dim=-1))
        x = F.dropout(x, DROPOUT, self.training)
        x, _ = self.lstm(x)
        x = F.dropout(x, DROPOUT, self.training)
        x = self.out_projection(x)
        return x
```

### **Wrapper**

```python
class VisualQuestionAnswering(L.LightningModule):
    def __init__(self):
        super().__init__()

        self.batch_size = BATCH_SIZE
        self.max_epoch = MAX_EPOCH
        self.lr = LEARNING_RATE
        self.lr_now = self.lr * 1e3

        self.image_extractor = VisionTransformer()

        # Patchify
        self.image_extractor.conv1.weight.data = clip.visual.conv1.weight.data.to(
            torch.float32
        ).clone()
        self.image_extractor.conv1.weight.requires_grad = UNFREEZE_PARAM

        # Class Embedding
        self.image_extractor.class_embedding.data = clip.visual.class_embedding.data.to(
            torch.float32
        ).clone()
        self.image_extractor.class_embedding.requires_grad = UNFREEZE_PARAM

        # Positional Embedding
        self.image_extractor.positional_embedding.data = (
            clip.visual.positional_embedding.data.to(torch.float32).clone()
        )
        self.image_extractor.positional_embedding.requires_grad = UNFREEZE_PARAM

        # Final Projection
        self.image_extractor.proj.data = clip.visual.proj.to(torch.float32).clone()
        self.image_extractor.proj.requires_grad = UNFREEZE_PARAM

        # Layer normalization before Transformer
        self.image_extractor.ln_pre.weight.data = clip.visual.ln_pre.weight.data.to(
            torch.float32
        ).clone()
        self.image_extractor.ln_pre.weight.requires_grad = UNFREEZE_PARAM

        self.image_extractor.ln_pre.bias.data = clip.visual.ln_pre.bias.data.to(
            torch.float32
        ).clone()
        self.image_extractor.ln_pre.bias.requires_grad = UNFREEZE_PARAM

        # Layer normalization after Transformer
        self.image_extractor.ln_post.weight.data = clip.visual.ln_post.weight.data.to(
            torch.float32
        ).clone()
        self.image_extractor.ln_post.weight.requires_grad = UNFREEZE_PARAM

        self.image_extractor.ln_post.bias.data = clip.visual.ln_post.bias.data.to(
            torch.float32
        ).clone()
        self.image_extractor.ln_post.bias.requires_grad = UNFREEZE_PARAM

        # Transformer
        for l, resblock in enumerate(self.image_extractor.transformer.resblocks):
            # Attention
            resblock.attn.out_proj.weight.data = (
                clip.visual.transformer.resblocks[l]
                .attn.out_proj.weight.data.to(torch.float32)
                .clone()
            )
            resblock.attn.out_proj.weight.requires_grad = UNFREEZE_PARAM

            resblock.attn.out_proj.bias.data = (
                clip.visual.transformer.resblocks[l]
                .attn.out_proj.bias.data.to(torch.float32)
                .clone()
            )
            resblock.attn.out_proj.bias.requires_grad = UNFREEZE_PARAM

            resblock.attn.in_proj_weight.data = (
                clip.visual.transformer.resblocks[l]
                .attn.in_proj_weight.data.to(torch.float32)
                .clone()
            )
            resblock.attn.in_proj_weight.requires_grad = UNFREEZE_PARAM

            resblock.attn.in_proj_bias.data = (
                clip.visual.transformer.resblocks[l]
                .attn.in_proj_bias.data.to(torch.float32)
                .clone()
            )
            resblock.attn.in_proj_bias.requires_grad = UNFREEZE_PARAM

            # 1st LayerNorm
            resblock.ln_1.weight.data = (
                clip.visual.transformer.resblocks[l]
                .ln_1.weight.data.to(torch.float32)
                .clone()
            )
            resblock.ln_1.weight.requires_grad = UNFREEZE_PARAM

            resblock.ln_1.bias.data = (
                clip.visual.transformer.resblocks[l]
                .ln_1.bias.data.to(torch.float32)
                .clone()
            )
            resblock.ln_1.bias.requires_grad = UNFREEZE_PARAM

            # 2nd LayerNorm
            resblock.ln_2.weight.data = (
                clip.visual.transformer.resblocks[l]
                .ln_2.weight.data.to(torch.float32)
                .clone()
            )
            resblock.ln_2.weight.requires_grad = UNFREEZE_PARAM

            resblock.ln_2.bias.data = clip.visual.transformer.resblocks[
                l
            ].ln_2.bias.data.to(torch.float32)
            resblock.ln_2.bias.requires_grad = UNFREEZE_PARAM

            # Attention Mask
            try:
                resblock.attn_mask = (
                    clip.visual.transformer.resblocks[l]
                    .attn_mask.to(torch.float32)
                    .clone()
                )
                resblock.attn_mask.requires_grad = UNFREEZE_PARAM
            except:
                pass

            # MLP
            resblock.mlp[0].weight.data = (
                clip.visual.transformer.resblocks[l]
                .mlp[0]
                .weight.data.to(torch.float32)
                .clone()
            )
            resblock.mlp[0].weight.requires_grad = UNFREEZE_PARAM

            resblock.mlp[0].bias.data = (
                clip.visual.transformer.resblocks[l]
                .mlp[0]
                .bias.data.to(torch.float32)
                .clone()
            )
            resblock.mlp[0].bias.requires_grad = UNFREEZE_PARAM

            resblock.mlp[2].weight.data = (
                clip.visual.transformer.resblocks[l]
                .mlp[2]
                .weight.data.to(torch.float32)
                .clone()
            )
            resblock.mlp[2].weight.requires_grad = UNFREEZE_PARAM

            resblock.mlp[2].bias.data = (
                clip.visual.transformer.resblocks[l]
                .mlp[2]
                .bias.data.to(torch.float32)
                .clone()
            )
            resblock.mlp[2].bias.requires_grad = UNFREEZE_PARAM

        # Token Embedding
        self.token_embedding = nn.Embedding(VOCAB_SIZE, HIDDEN_DIM)
        self.token_embedding.weight.data = clip.token_embedding.weight.data.to(
            torch.float32
        ).clone()
        self.token_embedding.weight.requires_grad = UNFREEZE_PARAM

        # Positional Embedding
        self.positional_embedding = nn.Parameter(torch.randn(MAX_SEQUENCE, HIDDEN_DIM))
        self.positional_embedding.data = clip.positional_embedding.data.to(
            torch.float32
        ).clone()
        self.positional_embedding.requires_grad = UNFREEZE_PARAM

        # Transformer
        self.transformer = Transformer(
            HIDDEN_DIM,
            12,
            8,
            build_attention_mask(),
        )
        for l, resblock in enumerate(self.transformer.resblocks):
            # Attention
            resblock.attn.out_proj.weight.data = (
                clip.transformer.resblocks[l]
                .attn.out_proj.weight.data.to(torch.float32)
                .clone()
            )
            resblock.attn.out_proj.weight.requires_grad = UNFREEZE_PARAM

            resblock.attn.out_proj.bias.data = (
                clip.transformer.resblocks[l]
                .attn.out_proj.bias.data.to(torch.float32)
                .clone()
            )
            resblock.attn.out_proj.bias.requires_grad = UNFREEZE_PARAM

            resblock.attn.in_proj_weight.data = (
                clip.transformer.resblocks[l]
                .attn.in_proj_weight.data.to(torch.float32)
                .clone()
            )
            resblock.attn.in_proj_weight.requires_grad = UNFREEZE_PARAM

            resblock.attn.in_proj_bias.data = (
                clip.transformer.resblocks[l]
                .attn.in_proj_bias.data.to(torch.float32)
                .clone()
            )
            resblock.attn.in_proj_bias.requires_grad = UNFREEZE_PARAM

            # 1st LayerNorm
            resblock.ln_1.weight.data = (
                clip.transformer.resblocks[l].ln_1.weight.data.to(torch.float32).clone()
            )
            resblock.ln_1.weight.requires_grad = UNFREEZE_PARAM

            resblock.ln_1.bias.data = (
                clip.transformer.resblocks[l].ln_1.bias.data.to(torch.float32).clone()
            )
            resblock.ln_1.bias.requires_grad = UNFREEZE_PARAM

            # 2nd LayerNorm
            resblock.ln_2.weight.data = (
                clip.transformer.resblocks[l].ln_2.weight.data.to(torch.float32).clone()
            )
            resblock.ln_2.weight.requires_grad = UNFREEZE_PARAM

            resblock.ln_2.bias.data = clip.transformer.resblocks[l].ln_2.bias.data.to(
                torch.float32
            )
            resblock.ln_2.bias.requires_grad = UNFREEZE_PARAM

            # Attention Mask
            try:
                resblock.attn_mask = (
                    clip.transformer.resblocks[l].attn_mask.to(torch.float32).clone()
                )
                resblock.attn_mask.requires_grad = UNFREEZE_PARAM
            except:
                pass

            # MLP
            resblock.mlp[0].weight.data = (
                clip.transformer.resblocks[l]
                .mlp[0]
                .weight.data.to(torch.float32)
                .clone()
            )
            resblock.mlp[0].weight.requires_grad = UNFREEZE_PARAM

            resblock.mlp[0].bias.data = (
                clip.transformer.resblocks[l].mlp[0].bias.data.to(torch.float32).clone()
            )
            resblock.mlp[0].bias.requires_grad = UNFREEZE_PARAM

            resblock.mlp[2].weight.data = (
                clip.transformer.resblocks[l]
                .mlp[2]
                .weight.data.to(torch.float32)
                .clone()
            )
            resblock.mlp[2].weight.requires_grad = UNFREEZE_PARAM

            resblock.mlp[2].bias.data = (
                clip.transformer.resblocks[l].mlp[2].bias.data.to(torch.float32).clone()
            )
            resblock.mlp[2].bias.requires_grad = UNFREEZE_PARAM

        # LayerNorm
        self.ln_final = nn.LayerNorm(HIDDEN_DIM)
        self.ln_final.weight.data = clip.ln_final.weight.data.to(torch.float32).clone()
        self.ln_final.bias.data = clip.ln_final.bias.data.to(torch.float32).clone()
        self.ln_final.weight.requires_grad = UNFREEZE_PARAM
        self.ln_final.bias.requires_grad = UNFREEZE_PARAM

        # Text Projection
        self.text_projection = nn.Parameter(torch.randn(HIDDEN_DIM, HIDDEN_DIM))
        self.text_projection.data = clip.text_projection.data.to(torch.float32).clone()
        self.text_projection.requires_grad = UNFREEZE_PARAM

        self.decoder = Decoder()

        self.automatic_optimization = False

        self.train_loss = list()
        self.val_loss = list()

        self.train_loss_recorder = AvgMeter()
        self.val_loss_recorder = AvgMeter()

        self.test_squad = SQuAD()

        self.sanity_check_counter = 1

    def encode_image(self, image):
        # shape = [*, width, grid, grid]
        image = self.image_extractor.conv1(
            image.type(self.image_extractor.conv1.weight.dtype)
        )
        # shape = [*, width, grid ** 2]
        image = image.reshape(image.shape[0], image.shape[1], -1)
        image = image.permute(0, 2, 1)  # shape = [*, grid ** 2, width]
        image = torch.cat(
            [
                self.image_extractor.class_embedding.to(image.dtype)
                + torch.zeros(
                    image.shape[0],
                    1,
                    image.shape[-1],
                    dtype=image.dtype,
                    device=image.device,
                ),
                image,
            ],
            dim=1,
        )  # shape = [*, grid ** 2 + 1, width]
        image = image + self.image_extractor.positional_embedding.to(image.dtype)
        image = self.image_extractor.ln_pre(image)
        image = image.permute(1, 0, 2)  # NLD -> LND
        image = self.image_extractor.transformer(image)
        image = image.permute(1, 0, 2)  # LND -> NLD

        image = self.image_extractor.ln_post(image[:, 0, :])
        image = image @ self.image_extractor.proj  # .to(image.dtype)

        return image

    def encode_text(self, text):
        # [batch_size, n_ctx, d_model]
        x = self.token_embedding(text)  # .type(torch.float16)
        x = x + self.positional_embedding.type(x.dtype)
        x = x.permute(1, 0, 2)  # NLD -> LND
        x = self.transformer(x.type(x.dtype))
        x = x.permute(1, 0, 2)  # LND -> NLD
        x = self.ln_final(x)
        x = x[
            torch.arange(x.shape[0]), text.argmax(dim=-1)
        ] @ self.text_projection.type(x.dtype)

        return x

    def encoder(self, image, question):
        image_feature = self.encode_image(image)
        question_feature = self.encode_text(question)

        # normalized features
        image_feature = image_feature / image_feature.norm(dim=1, keepdim=True)
        question_feature = question_feature / question_feature.norm(dim=1, keepdim=True)

        # shape: N, 1, D
        return image_feature.unsqueeze(1), question_feature.unsqueeze(1)

    def forward(self, image, question, answer):
        image_feature, question_feature = self.encoder(image, question)
        return self.decoder(answer, question_feature, image_feature)

    def respond(self, image, question):
        image_feature, question_feature = self.encoder(image, question)

        token = SimpleTokenizer.encode(START_TOKEN) + [0] * (MAX_SEQUENCE)
        for index in range(MAX_SEQUENCE):
            answer = torch.LongTensor(token[:-1]).unsqueeze(0).to(self.device)
            pred_token = self.decoder(answer, question_feature, image_feature)
            pred_token = pred_token.softmax(-1).argmax(2)[:, index].item()
            token[index + 1] = pred_token
            if pred_token == SimpleTokenizer.encode(END_TOKEN)[0]:
                break
        return (
            SimpleTokenizer.decode(token)
            .replace(START_TOKEN, "")
            .replace(END_TOKEN, "")
            .replace(PAD_TOKEN, "")
        )

    def training_step(self, batch, batch_nb):
        image, question, answer = batch

        pred = self(image, question, answer[:, :-1])
        pred = pred.view(-1, pred.shape[-1])
        answer = answer[:, 1:].reshape(answer.shape[0] * (answer.shape[-1] - 1))

        loss = F.cross_entropy(pred, answer, ignore_index=0)

        opt = self.optimizers()
        opt.zero_grad()
        self.manual_backward(loss)
        torch.nn.utils.clip_grad_value_(
            self.parameters(),
            math.sqrt(math.pi * math.e),
        )
        opt.step()

        self.log("train_loss", loss, prog_bar=True)
        self.train_loss_recorder.update(loss.data)

    def on_train_epoch_end(self):
        sch = self.lr_schedulers()
        sch.step()
        self.train_loss.append(self.train_loss_recorder.show().data.cpu().numpy())
        self.train_loss_recorder = AvgMeter()

    def validation_step(self, batch, batch_nb):
        image, question, answer = batch

        pred = self(image, question, answer[:, :-1])
        pred = pred.view(-1, pred.shape[-1])
        answer = answer[:, 1:].reshape(answer.shape[0] * (answer.shape[-1] - 1))

        loss = F.cross_entropy(pred, answer, ignore_index=0)

        if self.sanity_check_counter == 0:
            self.log("val_loss", loss, prog_bar=True)
            self.val_loss_recorder.update(loss.data)

    def on_validation_epoch_end(self):
        if self.sanity_check_counter == 0:
            loss = self.val_loss_recorder.show().data.cpu().numpy()
            lr_now_ = self.optimizers().param_groups[0]["lr"]
            if self.lr_now != lr_now_:
                self.lr_now = lr_now_
                str_report = f"[{MODEL_NAME}] Learning Rate Changed: {lr_now_}"
                str_report += f"- Epoch: {self.current_epoch}"
                print(str_report)
            self.val_loss.append(loss)
            self.val_loss_recorder = AvgMeter()
        else:
            self.sanity_check_counter -= 1

    def test_step(self, batch, batch_nb):
        image, question, answer = batch

        N_BATCH = image.shape[0]

        squad_pred = list()
        squad_target = list()

        for id in range(N_BATCH):
            res = self.respond(image[id].unsqueeze(0), question[id].unsqueeze(0))
            target = (
                SimpleTokenizer.decode(answer[id].cpu().detach().numpy().tolist())
                .replace(START_TOKEN, "")
                .replace(END_TOKEN, "")
                .replace(PAD_TOKEN, "")
            )
            squad_pred.append({"prediction_text": res, "id": id})
            squad_target.append(
                {
                    "answers": {"answer_start": [id], "text": [target]},
                    "id": id,
                }
            )

        squad = self.test_squad(squad_pred, squad_target)

        self.log("test_exact_match", squad["exact_match"], prog_bar=True, logger=True)
        self.log("test_f1", squad["f1"], prog_bar=True, logger=True)

    def on_train_end(self):
        # Loss
        loss_img_file = f"experiment/training/{MODEL_NAME}_loss_plot.png"
        plt.plot(self.train_loss, color="r", label="train")
        plt.plot(self.val_loss, color="b", label="validation")
        plt.title("Loss Curves")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.grid()
        plt.savefig(loss_img_file)
        plt.clf()
        img = cv2.imread(loss_img_file)
        cv2_imshow(img)

    def train_dataloader(self):
        return data.DataLoader(
            TrainDataset,
            batch_size=self.batch_size,
            shuffle=True,
            collate_fn=CollateFunction,
            num_workers=2,
            persistent_workers=True,
        )

    def val_dataloader(self):
        return data.DataLoader(
            ValDataset,
            batch_size=self.batch_size,
            shuffle=False,
            collate_fn=CollateFunction,
            num_workers=2,
            persistent_workers=True,
        )

    def test_dataloader(self):
        return data.DataLoader(
            TestDataset,
            batch_size=self.batch_size,
            shuffle=False,
            collate_fn=CollateFunction,
            num_workers=2,
            persistent_workers=True,
        )

    def configure_optimizers(self):
        optimizer = optim.AdamW(
            self.parameters(),
            self.lr,
            betas=(0.9, 0.98),
            eps=1e-9,
        )

        lr_scheduler = {
            "scheduler": optim.lr_scheduler.MultiStepLR(
                optimizer,
                milestones=[int(self.max_epoch * ms) for ms in MILESTONES],
                gamma=REDUCE_LR_FACTOR,
            ),
            "name": "lr_scheduler",
        }

        return [optimizer], [lr_scheduler]
```

## **Training**

```python
def _train_loop():
    seed_everything(SEED, workers=True)

    print(MODEL_NAME)
    model = MODEL()

    callbacks = list()

    checkpoint = ModelCheckpoint(
        monitor=METRIC_TO_MONITOR,
        dirpath=f"{EXPERIMENT_DIR}/model",
        mode=METRIC_MODE,
        filename=f"{MODEL_NAME}_best",
    )
    callbacks.append(checkpoint)

    early_stopping = EarlyStopping(
        monitor=METRIC_TO_MONITOR,
        min_delta=0.00,
        patience=EARLY_STOPPING_PATIENCE,
        verbose=False,
        mode=METRIC_MODE,
    )
    callbacks.append(early_stopping)

    if os.path.exists(BEST_MODEL_PATH):
        ckpt_path = BEST_MODEL_PATH
    else:
        ckpt_path = None

    trainer = Trainer(
        accelerator="auto",
        devices=1,
        max_epochs=MAX_EPOCH,
        logger=False,
        callbacks=callbacks,
        log_every_n_steps=5,
    )
    trainer.fit(model, ckpt_path=ckpt_path)

_train_loop()
```

```python
def _test_loop():
    trainer = Trainer(accelerator='auto', logger=False)
    model = MODEL()
    trainer.test(model=model, ckpt_path=BEST_MODEL_PATH)

_test_loop()
```
